I got this data from http://lib.stat.cmu.edu/database/visualizing.data.zip

```python
import csv 
f = open('environmental.csv', 'rb')
r = csv.reader(f)
temprature = []
wind = []
x=0
for row in r: 
    if x!=0:
        row=row[2:] 
        row=map(float,row) 
        temprature.append(row[0]) 
        wind.append(row[1]) 
    x+=1
f.close()   
from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('The relationship between the wind and the temprature')
ax.set_xlabel('wind_label')
ax.set_ylabel('temprature_label')
scatter(wind,temprature)
```

x axis represents the wind
y axis represents the temprature
The conclusion is that there's no relationship.
